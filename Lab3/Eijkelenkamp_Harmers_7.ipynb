# <span style="color:blue">Exercise 1 - Construct a generator network (40pt)  <a class="anchor" id="chapter1"></a></span>

The GAN generator learns to create fake data from random noises. It takes as input the noise vector of nz-dimensions and outputs an image of size $64\times 64$. The architecture of the generator consists of several ($<10$) tranposed convolutional blocks or convolutional blocks with upsampling operations. In the following, please provide the missing codes for generator. Briefly explain what are the outputs after every operation.



```python
# generator
class Generator(nn.Module):
    def __init__(self, nz):
        super(Generator, self).__init__()
        self.nz = nz # the input noise vector
        
        # First transposed convolution layer. The number of input channels is equal to the size of the latent noise vector,
        # and 1024 output channels are produced by the transposed convolution. Output: [batch_size, 1024, 4, 4]
        self.convT1 = nn.ConvTranspose2d(in_channels=nz, out_channels=image_size*16, kernel_size=4, stride=1, padding=0)
        # Batch normalization for first transposed convolution layer.
        self.bn1 = nn.BatchNorm2d(image_size*16)

        # Second transposed convolution layer. The number of input channels is equal to 1024,
        # and 512 output channels are produced by the transposed convolution. Output shape: [batch_size, 512, 8, 8]
        self.convT2 = nn.ConvTranspose2d(in_channels=image_size*16, out_channels=image_size*8, kernel_size=4, stride=2, padding=1)
        # Batch normalization for second transposed convolution layer.
        self.bn2 = nn.BatchNorm2d(image_size*8)

        # Third transposed convolution layer. The number of input channels is equal to 512,
        # and 256 output channels are produced by the transposed convolution. Output shape: [batch_size, 256, 16, 16]
        self.convT3 = nn.ConvTranspose2d(in_channels=image_size*8, out_channels=image_size*4, kernel_size=4, stride=2, padding=1)
        # Batch normalization for third transposed convolution layer.
        self.bn3 = nn.BatchNorm2d(image_size*4)

        # Fourth transposed convolution layer. The number of input channels is equal to 256,
        # and 128 output channels are produced by the transposed convolution. Output shape: [batch_size, 128, 32, 32]
        self.convT4 = nn.ConvTranspose2d(in_channels=image_size*4, out_channels=image_size*2, kernel_size=4, stride=2, padding=1)
        # Batch normalization for fourth transposed convolution layer.
        self.bn4 = nn.BatchNorm2d(image_size*2)

        # Last transposed convolution layer. The number of input channels is equal to 128,
        # and 3 output channels are produced by the transposed convolution, representing the RGB channels in a color image.
        # Output shape: [batch_size, 3, 64, 64]
        self.convT5 = nn.ConvTranspose2d(in_channels=image_size*2, out_channels=3, kernel_size=4, stride=2, padding=1)
        self.tanh = nn.Tanh()


    def forward(self, x):
      # 'convT1' layer
      x = F.leaky_relu(self.bn1(self.convT1(x)))
        
      # 'convT2' layer
      x = F.leaky_relu(self.bn2(self.convT2(x)))
        
      # 'convT3' layer
      x = F.leaky_relu(self.bn3(self.convT3(x)))

      # 'convT4' layer
      x = F.leaky_relu(self.bn4(self.convT4(x)))
        
      # 'convT5' layer
      x = self.tanh(self.convT5(x))
        
      return x
```

We have provided the missing codes for the GAN generator in the preceding code cell. The architecture of the generator is mostly based on the design of our neural network from the previous assignment, where we have switched out the 2D convolution layers with 2D *transposed* convolution layers and added an additional layer (i.e., this network has five layers in total as opposed to four in the previous assignment) in order to upsample the random noise vector to a RGB image of size $64 \times 64$ [1]. From [2], we know that the output height $H_{out}$ and output width $W_{out}$ can be computed with

$$H_{out}​ = (H_{in}​−1) \times stride[0]−2 \times padding[0] + dilation[0] \times (kernel\_size[0]−1) + output\_padding[0] + 1,$$

$$W_{out}​ = (W_{in}​−1) \times stride[1]−2 \times padding[1] + dilation[1] \times (kernel\_size[1]−1) + output\_padding[1] + 1.$$

Since we will only be dealing with square feature maps and kernels (and, if applicable, uniform padding on all sides), we have that the output height is equal to the output width. Furthermore, we have taken

$$dilation[0] = dilation[1] = 1$$

and 

$$output\_padding[0] = output\_padding[1] = 0$$

in all of our layers. Hence, the output height/width after every layer can be computed as follows:

$$\text{Layer 1}: W_{out}​ = (1​−1) \times 1 − 2 \times 0 + (4−1) + 1 = 4,$$

$$\text{Layer 2}: W_{out}​ = (4​−1) \times 2 − 2 \times 1 + (4−1) + 1 = 8,$$

$$\text{Layer 3}: W_{out}​ = (8−1) \times 2 − 2 \times 1 + (4−1) + 1 = 16,$$

$$\text{Layer 4}: W_{out}​ = (16​−1) \times 2 − 2 \times 1 + (4−1) + 1 = 32,$$

$$\text{Layer 5}: W_{out}​ = (32−1) \times 2 − 2 \times 1 + (4−1) + 1 = 64.$$

Note that the random noise vector has dimensions (100, 1, 1) such that the width and height of the first layer's input are equal to 1. Therefore, we have verified that the final layer's output has size $64 \times 64$, as desired. As suggested by [1], the number of output channels of the initial layer are set to an arbitrarily large number and reduced by a factor of 2 until the final layer is reached, where we assign three output channels (representing the three color channels in an RGB image). In between the layers, we have applied *LeakyReLU* as the activation functions to prevent neurons from 'dying' (as described in Lecture 4 of this course) and *Tanh* transforms the output of the final layer such that its values/coefficients are within the range of [-1, 1].  




# <span style="color:blue">Exercise 2 - Construct a discriminator network (40pt)  <a class="anchor" id="chapter2"></a></span>


The GAN discriminator learns to distinguish between the generated fake images and the real images. It takes as input an (fake or real) image and output its label (0 or 1). The architecture of the discriminator consists of several ($<10$) strided convolutional blocks or convolutional blocks with maxpooling operations. In the following, please provide the missing codes for discriminator. Briefly explain what are the expected outputs after every layer.


```python
# discriminator
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        
        # First convolution layer. The input channels are the three colour channels (RGB), 
        # and 64 output channels are produced by the convolution.
        # Output shape: [batch_size, 64, 32, 32]
        self.conv1 = nn.Conv2d(in_channels=3, out_channels=image_size, kernel_size=3, stride=1, padding=1)
        # Batch normalization for first convolution layer.
        self.bn1 = nn.BatchNorm2d(image_size)
        # Max pooling for first convolution layer.
        self.maxpool1 = nn.MaxPool2d(kernel_size=2)
        
        # Second convolution layer. The number of input channels is equal to 64,
        # and 128 output channels are produced by the convolution. Output shape: [batch_size, 128, 16, 16]
        self.conv2 = nn.Conv2d(in_channels=image_size, out_channels=image_size*2, kernel_size=3, stride=2, padding=1)
        # Batch normalization for second convolution layer.
        self.bn2 = nn.BatchNorm2d(image_size*2)
        
        # Third convolution layer. The number of input channels is equal to 128,
        # and 256 output channels are produced by the convolution. Output shape: [batch_size, 256, 8, 8]
        self.conv3 = nn.Conv2d(in_channels=image_size*2, out_channels=image_size*4, kernel_size=3, stride=2, padding=1)
        # Batch normalization for third convolution layer.
        self.bn3 = nn.BatchNorm2d(image_size*4)

        # Fourth convolution layer. The number of input channels is equal to 256,
        # and 512 output channels are produced by the convolution. Output shape: [batch_size, 512, 4, 4]
        self.conv4 = nn.Conv2d(in_channels=image_size*4, out_channels=image_size*8, kernel_size=3, stride=2, padding=1)
        # Batch normalization for fourth convolution layer.
        self.bn4 = nn.BatchNorm2d(image_size*8)
        
        # Last convolution layer. The number of input channels is equal to 512,
        # and 1 output channel is produced by the convolution (in accordance with the image label of 0 or 1, size 1 by 1). 
        # Output shape: [batch_size, 1, 1, 1]
        self.conv5 = nn.Conv2d(in_channels=image_size*8, out_channels=1, kernel_size=3, stride=2, padding=0)
        self.sigmoid = nn.Sigmoid()
        
        
    def forward(self, x):
        # 'conv1' layer
        x = F.leaky_relu(self.bn1(self.conv1(x)))
        x = self.maxpool1(x)
        
        # 'conv2' layer
        x = F.leaky_relu(self.bn2(self.conv2(x)))
        
        # 'conv3' layer
        x = F.leaky_relu(self.bn3(self.conv3(x)))

        # 'conv4' layer
        x = F.leaky_relu(self.bn4(self.conv4(x)))

        # 'conv5' layer
        x = self.sigmoid(self.conv5(x))
        
        return x
```

We have provided the missing codes for the GAN discriminator in the preceding code cell. The architecture of the discriminator is mostly based on the design of our neural network from the previous assignment, where we have now used five 2D convolution layers (as opposed to four in the previous assignment) in order to downsample an RGB image of size $64 \times 64$ to an 'image' of size $1 \times 1$. From [3], we know that the output height $H_{out}$ and output width $W_{out}$ can be computed with

$$H_{out}​ = \left\lfloor \frac{H_{in} + 2 \times padding[0] - dilation[0] \times (kernel\_size[0] - 1) - 1}{stride[0]} + 1 \right\rfloor,$$

$$W_{out}​ = \left\lfloor \frac{W_{in} + 2 \times padding[1] - dilation[1] \times (kernel\_size[1] - 1) - 1}{stride[1]} + 1 \right\rfloor.$$

Since we will only be dealing with square feature maps and kernels (and, if applicable, uniform padding on all sides), we have that the output height of the feature maps is equal to their output width. Furthermore, we have taken

$$dilation[0] = dilation[1] = 1$$

in all of our layers. Hence, the output height/width after every layer can be computed as follows:

$$\text{Layer 1}: W_{out}​ = \left\lfloor \frac{64 + 2 \times 1 - (3 - 1) - 1}{1} + 1 \right\rfloor = 64$$

$$\text{Max Pooling}: W_{out} = \frac{W_{in}}{2} = \frac{64}{2} = 32$$

$$\text{Layer 2}: W_{out}​ = \left\lfloor \frac{32 + 2 \times 1 - (3 - 1) - 1}{2} + 1 \right\rfloor = 16$$

$$\text{Layer 3}: W_{out}​ = \left\lfloor \frac{16 + 2 \times 1 - (3 - 1) - 1}{2} + 1 \right\rfloor = 8$$

$$\text{Layer 4}: W_{out}​ = \left\lfloor \frac{8 + 2 \times 1 - (3 - 1) - 1}{2} + 1 \right\rfloor = 4$$

$$\text{Layer 5}: W_{out}​ = \left\lfloor \frac{4 + 2 \times 0 - (3 - 1) - 1}{2} + 1 \right\rfloor = 1$$

Therefore, we have verified that the final layer's output has size $1 \times 1$, as desired. Note that when the final layer is reached, we assign one output channel in accordance with the output label of size $1 \times 1$. In between the layers, we have again applied *LeakyReLU* as the activation functions to prevent neurons from 'dying' (as described in Lecture 4) and *Sigmoid* transforms the output of the final layer to a value in the range [0, 1].  

# <span style="color:blue">Exercise 3 - Train your GAN (20pt)  <a class="anchor" id="chapter3"></a></span>


Train the network and report the obtained loss (both generator and discriminator) plot with respect to the epochs and the generated images.


```python
# initialize models
generator = Generator(nz).to(device)
discriminator = Discriminator().to(device)

# optimizers
optim_g = optim.Adam(generator.parameters(), lr=lr, betas=(beta1, 0.999))
optim_d = optim.Adam(discriminator.parameters(), lr=lr, betas=(beta1, 0.999))

# loss function
criterion = nn.BCELoss()

# to store the generator and discriminator loss after each epoch
losses_g = []
losses_d = []

# creat the noise signal as input to the generator
def create_noise(sample_size,nz):
  return torch.randn(sample_size,nz,1,1).to(device)

# function to train the discriminator network
def train_discriminator(optimizer, data_real, data_fake):
    b_size = data_real.size(0)
    # get the real label vector
    real_label = torch.ones(b_size, 1).to(device).squeeze()
    # get the fake label vector
    fake_label = torch.zeros(b_size, 1).to(device).squeeze()
    optimizer.zero_grad()
    # get the outputs by doing real data forward pass
    output_real = discriminator(data_real).view(-1)
    loss_real = criterion(output_real, real_label)
    # get the outputs by doing fake data forward pass
    output_fake = discriminator(data_fake).view(-1)
    loss_fake = criterion(output_fake, fake_label)
    # compute gradients of real loss 
    loss_real.backward()
    # compute gradients of fake loss
    loss_fake.backward()
    # update discriminator parameters
    optimizer.step()
    return loss_real + loss_fake


# function to train the generator network
def train_generator(optimizer, data_fake):
    b_size = data_fake.size(0)
    # get the real label vector
    real_label = torch.ones(b_size, 1).to(device).squeeze()
    optimizer.zero_grad()
    # output by doing a forward pass of the fake data through discriminator
    output = discriminator(data_fake).view(-1)
    loss = criterion(output, real_label)
    # compute gradients of loss
    loss.backward()
    # update generator parameters
    optimizer.step()
    return loss 

# create the noise vector
noise = create_noise(sample_size, nz)

for epoch in range(epochs):
    loss_g = 0.0
    loss_d = 0.0
    for batch_idx, data in enumerate(trainloader):
        image, _ = data
        image = image.to(device)
        b_size = len(image)
        # forward pass through generator to create fake data
        data_fake = generator(create_noise(b_size, nz)).detach()
        data_real = image
        loss_d += train_discriminator(optim_d, data_real, data_fake)
        data_fake = generator(create_noise(b_size, nz))
        loss_g += train_generator(optim_g, data_fake)
    # final forward pass through generator to create fake data...
    # ...after training for current epoch
    generated_img = generator(noise).cpu().detach()
    # save the generated torch tensor models to disk
    
    # We added this line to save the images in the correct Google Drive folder.
    path = root_path+f"outputs/gen_img{epoch}.png"   # We added this line to save the images in the correct Google Drive folder.
    
    save_image(generated_img, path, normalize=True)
    epoch_loss_g = loss_g / batch_idx # total generator loss for the epoch
    epoch_loss_d = loss_d / batch_idx # total discriminator loss for the epoch
    losses_g.append(epoch_loss_g)
    losses_d.append(epoch_loss_d)
    print(f"Epoch {epoch+1} of {epochs}")
    print(f"Generator loss: {epoch_loss_g:.8f}, Discriminator loss: {epoch_loss_d:.8f}")
```

```python
# Plotting the generator and discriminator loss with respect to the epochs.

gen_loss = [losses_g[i].item() for i in range(0, epochs)]   # generator loss
dis_loss = [losses_d[i].item() for i in range(0, epochs)]   # discriminator loss

fig  = plt.figure(figsize=[8, 8])
frame = fig.add_subplot(1, 1, 1)
frame.plot(range(0, epochs, 1), gen_loss, label="Generator loss")
frame.plot(range(0, epochs, 1), dis_loss, label="Discriminator loss")
frame.legend()
frame.set_xlabel("Epoch")
frame.set_ylabel("Loss")
frame.set_title("Plot of Generator and Discriminator Loss throughout the Epochs")

plt.show()
```
